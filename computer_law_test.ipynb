```python
import hashlib
import os
import difflib
```

```python

# Function to read the file and return its contents
def read_file(filepath):
    with open(filepath, 'r', errors='ignore') as file:
        return file.read()

# Function to compare the similarity of two files
def compare_files(file1, file2):
    content1 = read_file(file1)
    content2 = read_file(file2)
    
    matcher = difflib.SequenceMatcher(None, content1, content2)

    return matcher  # This returns a float between 0 and 1, representing the percentage of similarity

# Function to recursively list all files in a directory
def list_files(directory):
    return [os.path.join(dp, f) for dp, dn, filenames in os.walk(directory) for f in filenames if f.endswith('.java')]

# Function to compare all files between two directories
def compare_directories(dir1, dir2, threshold=0.09):
    files_dir1 = list_files(dir1)
    files_dir2 = list_files(dir2)

    similarities = []
    file1s = []
    file2s = []
    for file1 in files_dir1:
        for file2 in files_dir2:
            comparison = compare_files(file1, file2)
            similarity = comparison.ratio()
            file1s.append(file1)
            file2s.append(file2)


            similarities.append(similarity)
            if similarity > threshold:  # Only consider files with similarity above the threshold
                print(comparison.get_opcodes())

    similarity_dict = {}

    for score, str1, str2 in zip(similarities, file1s, file2s):

        # Check if the score is a key in the dictionary
        if score in similarity_dict:
            similarity_dict[score].append((str1, str2))
        else:
            similarity_dict[score] = [(str1, str2)]

    # Dictionary has similarity scores as keys; each key points to a list of string pairs

    similarities.sort(reverse=True)
    sub = similarities[10:30]
    for i in sub:
        print(i)

    return similarity_dict
```

```python
# Directories
dir1 = './core'
dir2 = './debugger'
diction = compare_directories(dir1, dir2)
```
